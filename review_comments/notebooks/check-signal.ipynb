```python
rdf_signal = load_signal_rdf()
df_signal = load_df(rdf_signal, 1)
```

```python
def add_m(df_signal, i,j):
    df_signal[f'm{i}{j}'] =   np.sqrt( np.square(df_signal[f'p{i}_fromXc_PE'] + df_signal[f'p{j}_fromXc_PE'])  \
                             - np.square( df_signal[f'p{i}_fromXc_PX'] + df_signal[f'p{j}_fromXc_PX']) \
                             - np.square( df_signal[f'p{i}_fromXc_PY'] + df_signal[f'p{j}_fromXc_PY']) \
                             - np.square( df_signal[f'p{i}_fromXc_PZ'] + df_signal[f'p{j}_fromXc_PZ']))
```

```python
add_m(df_signal, 1, 2)
add_m(df_signal, 2, 3)
add_m(df_signal, 1, 3)
```
