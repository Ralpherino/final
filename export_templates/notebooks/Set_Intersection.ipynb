```python
part1 = [['24'], ['13', '7', '27', '15', '29', '16', '23', '21', '26', '18', '25', '14', 'others', '19', '22', '20', '0']]
part2 = [['20'], ['29', '18'], ['21'], ['25', '14', '13', '7', '19', '16', '24'], ['27', 'others', '15', '23', '26', '22', '0']]
part3 = [['14', '15', '16', '21', '24'], ['others'], ['20', '0'], ['13', '7', '27', '29', '23', '26', '18', '25', '19', '22']]

```

```python
elements = [ e for s in part1 for e in s ] 
```

```python
final = set()
for e in elements:
    if len([ p for p in final if e in p ])  == 0:
        print(f"Finding set for {e}")
        p1 = [ p for p in part1 if e in p ][0]
        p2 = [ p for p in part2 if e in p ][0]
        p3 = [ p for p in part3 if e in p ][0]
        print(p1)
        print(p2)
        print(p3)
        i = set(p1).intersection(set(p2)).intersection(set(p3))
        print(f"Adding {i}")
        final.add(frozenset(i))
    else:
        print(f"{e} already in {[ p for p in final if e in p ]}")
```
