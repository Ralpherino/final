```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
%matplotlib inline 
```

## Loading the file

```python
datatype = "2012"
eventtype = "13763200"
polarity = "magdown"
sign = "None"
data_prefix = "root://eoslhcb.cern.ch//eos/lhcb/wg/semileptonic/RDsHad/AP/pidgen_merged/"
filename =  data_prefix + f"rds_final_{datatype}_{eventtype}_{polarity}_{sign}.root"
```

```python
print(f"Loading DecayTree from {filename}")
rdn = ROOT.RDataFrame("DecayTree", [ filename ])
print(f"Number of candidates: {rdn.Count().GetValue()}")
```

## Histograms

```python
ctxt = []

def plot(var, hnbin, hmin, hmax, context, normalize=False, ylogscale=False):
    c = ROOT.TCanvas()
    hn = rdn.Histo1D((f"New {var}", f"{var}", hnbin, hmin, hmax), var)
    if normalize:
        hn.Scale(1./hn.GetEntries())
    if ylogscale:
        c.SetLogy()
    hn.Draw("")
    c.Draw()
    context += [c, hn ]
    
```
