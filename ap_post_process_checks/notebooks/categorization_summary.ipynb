# Categorization of the candidates in the Hadronic RDs MC samples

```python
import ROOT
import numpy as np
import matplotlib.pyplot as plt
import os
import pandas as pd
import json
from XRootD import client
from IPython.display import display, HTML
pd.options.display.latex.repr = True
%matplotlib inline 
```

## List of event types and utility methods

```python
## Should get this from the config
#data_prefix = "root://eoslhcb.cern.ch//eos/lhcb/wg/semileptonic/RDsHad/AP/pidgen_merged/"
data_prefix = "root://eoslhcb.cern.ch//eos/lhcb/user/b/bcouturi/rds/pidgen_merged"

# eventtypes for which there is no right sign/wrong sign samples
eventtypes_nosign = [ "13763200", "13266069", "13863401", "11266009"]

# Inclusive MC event types
eventtypes_inclMC = [ "23903003", "23903000",]

# Inclusive MC event types
eventtypes_signal = [ "13563002", ]

# eventtypes for which we have right sign/wrong sign samples
eventtypes_sign = eventtypes_signal + eventtypes_inclMC

# other attributes of the dataset
polarities = ["magup", "magdown"]
datatypes = ["2012"]
signs = ["ws", "rs"]

def get_filenames(datatype, eventtype, sign):
        filenames = [ f"{data_prefix}/rds_categ_{datatype}_{eventtype}_{polarity}_{sign}.root" 
                      for polarity in polarities ]
        return filenames
```

```python
# Loading the list of categories 
metadata_filename =  data_prefix + "/rds_categ_metadata_2012_23903003_magdown_rs.json"
with client.File() as f:
    f.open(metadata_filename)
    buffer = f.read()
    metadata = json.loads(buffer[1])
detailed_categories_map = metadata["detailed_categories"]
simple_categories_map = metadata["simple_categories"]
```

```python
def mygroupby(d, groupbycols):
    g = d.groupby(groupbycols).size().reset_index(name='count').sort_values([ 'count'], ascending=False).reset_index(drop=True)
    g["Percentage"] = g.apply(lambda row: 100 * row["count"]/d.shape[0], axis=1)
    g["cumulative %"] = g["Percentage"].cumsum(axis = 0)
    return g
```

```python
filter_vars = [ "simplified", "category"]
plot_vars = [ "B_Y_SEP", "B_M", "Y_M", "q2_2", "tauY_2" ]
```

```python
def get_dataframe(datatype, eventtype, sign, filter=None):
    filenames =  get_filenames(datatype, eventtype, sign)
    print(filenames)
    rdf = ROOT.RDataFrame("DecayTree", filenames)
    if filter is not None:
        rdf = rdf.Filter(filter)
    df = pd.DataFrame(rdf.Cache(plot_vars + filter_vars).AsNumpy())
    return df
```

```python
def print_stats(datatype, eventtype, sign, filter=None):
    df = get_dataframe(datatype, eventtype, sign, filter)
    df["category name"] = df.apply(lambda row: detailed_categories_map[str(int(row['category']))], axis=1)
    df["simplified category"] = df.apply(lambda row: simple_categories_map[str(int(row['simplified']))], axis=1)
    g = mygroupby(df, ['category name'])
    display(HTML(g.to_html()))
    gs = mygroupby(df, ['simplified category'])
    display(HTML(gs.to_html()))
```
