```python
def mygroupby(d, groupbycols):
    """ Utility to group by a column and show relative statistics """
    g = d.groupby(groupbycols).size().reset_index(name='count').sort_values([ 'count'], ascending=False).reset_index(drop=True)
    g["Percentage"] = g.apply(lambda row: 100 * row["count"]/d.shape[0], axis=1)
    g["cumulative %"] = g["Percentage"].cumsum(axis = 0)
    return g
```

## Reweighting (or scaling ? to be decided) the sample based on the nSPDHits

```python
h_mc, bins = np.histogram(mc_norm_cut.nSPDHits, range=[0, 800], bins=200)
h_data, bins = np.histogram(data_norm_cut.nSPDHits, range=[0, 800], bins=200)
```

```python

bincenters = (bins[1:] + bins[:-1])/2
xdata = bins[12:150]
ydata = (h_data/ h_mc)[12:150]
plt.plot( xdata , ydata);
```

```python
from scipy.optimize import curve_fit

def fitfunc(x, a, b, c):
    return a + b * x**2 + c *x * x * x

popt, pcov = curve_fit(fitfunc, xdata, ydata)
```

```python
plt.plot( xdata , ydata, label="Data");
plt.plot(xdata, fitfunc(xdata, *popt), 'g--', label='fit: a=%5.3e, b=%5.3e, c=%5.3e' % tuple(popt))
plt.legend();
```

```python
mc_norm_cut['nSPDHits_weight'] = mc_norm_cut.apply(lambda row: fitfunc(row.nSPDHits, *popt), axis=1)
mc_norm['nSPDHits_weight'] = mc_norm.apply(lambda row: fitfunc(row.nSPDHits, *popt), axis=1)
```

## Reweighting the sample based on the Y_M mass

```python
plot("Y_M")
```

```python
myrange=[600, 3400]
mybins=100
h_mc, bins = np.histogram(mc_norm_cut.Y_M, range=myrange, bins=mybins)
h_data, bins = np.histogram(data_norm_cut.Y_M, range=myrange, bins=mybins)
bincenters = (bins[1:] + bins[:-1])/2
xdata = bincenters
```

```python
ratio = np.divide(h_data, h_mc, where=(h_mc != 0))
ydata = ratio
# Compute error on ratio (null if cannot be computed)
error = np.divide(h_data * np.sqrt(h_mc) + h_mc * np.sqrt(h_data),
                  np.power(h_mc, 2),
                  where=(h_mc != 0))
plt.plot(bincenters, ratio)
plt.errorbar(bincenters, ratio, yerr=error, fmt='.', color='r')
```

```python
# from scipy.optimize import curve_fit
# def ymfitfunc(x, a, b, c, d):
#      return a * np.exp( (x-b)**2/c**2) + d
# popt, pcov = curve_fit(ymfitfunc, xdata, ydata, sigma=error)

weights = np.divide(1, error**2, where=(error != 0))
weights[0] = 1 # large error as we do no care much for the fit
weights = weights / sum(weights)

p = np.polynomial.Chebyshev.fit(xdata, ydata, 10, w=weights)

from scipy.interpolate import make_smoothing_spline
spl = make_smoothing_spline(xdata, ydata, lam=500, w=weights)
```

```python
# weights = np.nan_to_num(weights, posinf=0)
# plt.plot(xdata, weights)

plt.plot( xdata , ydata, label="Data");
#plt.plot(xdata, ymfitfunc(xdata, *popt), label="fit")
plt.plot(xdata, p(xdata), label="Cheb Fit")

plt.plot(xdata, spl(xdata), label="Spline Fit")

#plt.errorbar(bincenters, ratio, yerr=error, fmt='.', color='r')
plt.legend();
```

```python
def ymfit(x):
    if x < myrange[0] or x > myrange[1]:
        return 0
    else:
        return p(x)
```

```python
mc_norm_cut['Y_M_weight'] = mc_norm_cut.apply(lambda row: ymfit(row.Y_M), axis=1)
mc_norm['Y_M_weight'] = mc_norm.apply(lambda row: ymfit(row.Y_M), axis=1)
```
