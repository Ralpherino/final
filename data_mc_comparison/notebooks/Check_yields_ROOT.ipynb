# Inclusive MC files without the B_Y_SEP cut

```python
files = ['root://eoslhcb.cern.ch//eos/lhcb/wg/semileptonic/RDsHad/AP/final/ap_post_process/validated/rds_finalnoBYsep_2012_23903000_magdown_rs.root', 
         'root://eoslhcb.cern.ch//eos/lhcb/wg/semileptonic/RDsHad/AP/final/ap_post_process/validated/rds_finalnoBYsep_2012_23903000_magup_rs.root']
```

```python
rdf = ROOT.RDataFrame("DecayTree", files)
```

```python
rdf.Count().GetValue()
```

```python
rdf_categ = categories4.add_categories_and_filter(rdf,  apply_BDT_Iso_cut=False, apply_PIDK_cut=False)
```

```python
{ k:v for k,v in categories4.categories_map.items() if v == 'Xc_signal_Ypis_B_vertex_fromBs' }
```

## Applying the cuts to the RDataFrame

```python
%%time
rdf2 = rdf_categ.Filter("B_M > 5000", "BMsup5000")
rdf2 = rdf2.Filter("BDT_Iso > 0.03458", "BDTIsocut")
rdf2 = rdf2.Filter("Xc_Selection>0", "Xc_Selection")
rdf2 = rdf2.Filter("B_M > 5346 && B_M < 5386", "BM_near_peak")
rdf2 = rdf2.Filter("category == 2", "Xc_signal_Ypis_B_vertex_fromBs")
rdf2 = rdf2.Filter("abs(mN2v)<250e6", "mN2v")
rep = rdf2.Report()
rep.Print()
```
