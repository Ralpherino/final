```python
import matplotlib.pyplot as plt
import pandas as pd
```

```python
df = pd.read_csv("../report.pdf.csv")
```

```python
df
```

```python
df.index = df["params"]
dfp = df[["params", "mean_test_score", "std_test_score"]].sort_values(by="mean_test_score").head(20)
dfp[["params", "mean_test_score", "std_test_score"]].plot( kind="barh", legend=False, label="params", xerr="std_test_score")
plt.xlim([0.80, .93])

```

```python
dfp = df[["params", "mean_train_score", "std_train_score"]].sort_values(by="mean_train_score").head(20)
dfp[["params", "mean_train_score", "std_train_score"]].plot( kind="barh", legend=False, label="params", xerr="std_train_score")
plt.xlim([0.80, .93])

```
