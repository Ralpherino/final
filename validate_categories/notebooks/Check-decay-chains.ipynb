```python
import joblib
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import categories4
pd.set_option('display.min_rows', 50)
pd.set_option('display.max_columns', 500)
pd.set_option('max_colwidth', None)
```

```python
def mygroupby(d, groupbycols):
    g = d.groupby(groupbycols).size().reset_index(name='count').sort_values([ 'count'], ascending=False).reset_index(drop=True)
    g["Percentage"] = g.apply(lambda row: 100 * row["count"]/d.shape[0], axis=1)
    g["cumulative %"] = g["Percentage"].cumsum(axis = 0)
    return g
```

```python
def load_data(categ):
    df = joblib.load(f"../rds_categdecays_{categ}_2012_23903000_magdown_rs.pkl")
    dfd = df.query("Xc_Selection == 1 & B_Y_SEP < -4.5 & q2_2>0 & BDT_Iso > 0.03458 & B_MM < 5000")
    return dfd
```

```python
print(f"Categorie 14: {categories4.categories_map[14]}")
df = load_data(14)
g = mygroupby(df[['decay']], 'decay')
g
```
