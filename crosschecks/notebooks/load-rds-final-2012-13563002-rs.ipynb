```python
# Setting up the python/jupyter environment
import pandas as pd
import ROOT
import matplotlib.pyplot as plt
import numpy as np
from IPython.display import Image
import tabulate

%matplotlib inline
plt.rcParams["figure.figsize"] = (20,10)
from IPython.display import display, HTML
display(HTML("<style>.container { width:95% !important; }</style>"))

import warnings
warnings.filterwarnings("ignore", category=DeprecationWarning)
```

```python
import json
import os
from pathlib import Path
with open(Path(os.getcwd()) / ".." / "config" / "config.json") as f:
    cfg = json.load(f)
data_prefix = cfg["DATA_DIR"]
print(f"Loading data from {data_prefix}")


def get_filenames(datatype, eventtype, sign):
    polarities = [ "magdown", "magup" ]
    filenames = [ f"{data_prefix}/rds_final_{datatype}_{eventtype}_{polarity}_{sign}.root" 
                      for polarity in polarities ]
    return filenames
filenames = get_filenames("2012", "13563002", "rs")
rdf = ROOT.RDataFrame("DecayTree", filenames)
```
